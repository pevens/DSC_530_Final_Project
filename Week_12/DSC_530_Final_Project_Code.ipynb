```python
import thinkplot
import thinkstats2

# Create High and Low Budget DataFrames
high_budget = anime_df[anime_df['Budget (in Million USD)'] >= 100]
low_budget = anime_df[anime_df['Budget (in Million USD)'] < 100]

# Create PMFs
high_budget_pmf = thinkstats2.Pmf(high_budget['Ratings'])
low_budget_pmf = thinkstats2.Pmf(low_budget['Ratings'])

# Labels
low_budget_pmf.label = 'Budget < 100M'
high_budget_pmf.label = 'Budget >= 100M'

# Plotting PMFs as histogram
thinkplot.PrePlot(2, cols=2)
thinkplot.Hist(high_budget_pmf, align='right', width=.05)
thinkplot.Hist(low_budget_pmf, align='left', width=.05)
thinkplot.Config(xlabel='Ratings',
ylabel='probability',
axis = [6, 10.2, 0, 0.2],
legend=True)

# Plotting PMFs as step function
thinkplot.PrePlot(2)
thinkplot.SubPlot(2)
thinkplot.Pmfs([high_budget_pmf, low_budget_pmf])
thinkplot.Show(xlabel='Ratings',
ylabel='probability',
axis =[6, 10.2, 0, 0.2], 
legend=True)
```

```python
def MakeNormalPlot(budget):
    mean = budget.mean()
    std = budget.std()
    xs = [-3, 3]
    fxs, fys = thinkstats2.FitLine(xs, inter=mean, slope=std)

    thinkplot.Plot(fxs, fys, color='gray', label='model')
    xs, ys = thinkstats2.NormalProbability(budget)
    thinkplot.Plot(xs, ys, label='Budget in Million USD')
    thinkplot.Config(title='Normal Probability Plot',
                 xlabel='Standard deviations from mean',
                 ylabel='Budget in Million USD')

# Create Normal Probability Plot
MakeNormalPlot(anime_df['Budget (in Million USD)'].sort_values())
```

```python
import math

# Covariance
def Cov(xs, ys, meanx=None, meany=None):
    xs = np.asarray(xs)
    ys = np.asarray(ys)
    if meanx is None:
        meanx = np.mean(xs)
    if meany is None:
        meany = np.mean(ys)
    cov = np.dot(xs-meanx, ys-meany) / len(xs)
    return cov

# Pearsons Correlation
def Corr(xs, ys):
    xs = np.asarray(xs)
    ys = np.asarray(ys)
    meanx, varx = thinkstats2.MeanVar(xs)
    meany, vary = thinkstats2.MeanVar(ys)
    corr = Cov(xs, ys, meanx, meany) / math.sqrt(varx * vary)
    return corr
# Spearmans Correlation
def SpearmanCorr(xs, ys):
    xranks = pd.Series(xs).rank()
    yranks = pd.Series(ys).rank()
    return Corr(xranks, yranks)
```

Hypothesis Testing

H0 = Production budget has no effect on anime ratings for the top 100 anime.

Ha = Production budget has a positive effect on anime ratings for the top 100 anime.


```python
class CorrelationPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        xs, ys = data
        test_stat = abs(thinkstats2.Corr(xs, ys))
        return test_stat

    def RunModel(self):
        xs, ys = self.data
        xs = np.random.permutation(xs)
        return xs, ys
    

ht_anime_df = anime_df.dropna(subset=['Budget (in Million USD)', 'Ratings'])
data = ht_anime_df['Budget (in Million USD)'].values, anime_df['Ratings'].values
# pvalue = thinkstats2.HypothesisTest.PValue(data)

ht = CorrelationPermute(data)
pvalue = ht.PValue()
print(f"P-Value: ", pvalue)

print(f"Actual Correlation: ", ht.actual, "\nLargest Simulated Value: ", ht.MaxTestStat())
```
