```python
import pandas as pd

# Load data from a CSV file
anime_df = pd.read_csv(r"C:\Users\Skylar\Downloads\most_watched_anime_dataset_100_entries.csv")
print(anime_df)
```

```python
import matplotlib.pyplot as plt

# Plotting a Number of Episodes histogram
plt.hist(anime_df['Number of Episodes'], bins=10, color='skyblue', edgecolor='black')
 
# Adding labels and title
plt.xlabel('Number of Episodes')
plt.ylabel('Frequency')
plt.title('Number of Episodes Histogram')
 
# Display the plot
plt.show()
```

```python
# Plotting Ratings histogram
plt.hist(anime_df['Ratings'], bins=10, color='skyblue', edgecolor='black')
 
# Adding labels and title
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.title('Ratings Histogram')
 
# Display the plot
plt.show()
```

```python
# Plotting Budget histogram
plt.hist(anime_df['Budget (in Million USD)'], bins=20, color='skyblue', edgecolor='black')
 
# Adding labels and title
plt.xlabel('Budget (in Million USD)')
plt.ylabel('Frequency')
plt.title('Budget Histogram')
 
# Display the plot
plt.show()
```

```python
# Plotting Release Year histogram
plt.hist(anime_df['Release Year'], bins=32, color='skyblue', edgecolor='black')
 
# Adding labels and title
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.title('Release Year Histogram')
 
# Display the plot
plt.show()
```

```python
# Plotting Episode Duration histogram
plt.hist(anime_df['Duration per Episode (minutes)'], bins=12, color='skyblue', edgecolor='black')
 
# Adding labels and title
plt.xlabel('Episode Duration (in minutes)')
plt.ylabel('Frequency')
plt.title('Average Episode Duration Histogram')
 
# Display the plot
plt.show()
```

```python
# Variable Descriptive Statistics
mean = anime_df['Duration per Episode (minutes)'].mean()
var = anime_df['Duration per Episode (minutes)'].var()
std = anime_df['Duration per Episode (minutes)'].std()
skew1 = (mean - 30) / std
skew2 = (mean - 56) / std

print(mean)
print(var)
print(std)
print(skew1)
print(skew2)

# Count of each unique value in the 'Duration per Episode (minutes)' column
print(anime_df['Duration per Episode (minutes)'].value_counts())
```

```python
# Variable Descriptive Statistics
mean = anime_df['Number of Episodes'].mean()
var = anime_df['Number of Episodes'].var()
std = anime_df['Number of Episodes'].std()
skew = (mean - 69) / std

print(mean)
print(var)
print(std)
print(skew)

# Count of each unique value in the 'Duration per Episode (minutes)' column
print(anime_df['Number of Episodes'].value_counts())
```

```python
# Variable Descriptive Statistics
mean = anime_df['Ratings'].mean()
var = anime_df['Ratings'].var()
std = anime_df['Ratings'].std()
skew = (mean - 9.4) / std

print(mean)
print(var)
print(std)
print(skew)
# Count of each unique value in the 'Duration per Episode (minutes)' column
print(anime_df['Ratings'].value_counts())
```

```python
# Variable Descriptive Statistics
mean = anime_df['Budget (in Million USD)'].mean()
median = anime_df['Budget (in Million USD)'].median()
var = anime_df['Budget (in Million USD)'].var()
std = anime_df['Budget (in Million USD)'].std()
skew = 3 * (mean - median) / std # Pearsons second coefficient

print(mean)
print(median)
print(var)
print(std)
print(skew)

# Count of each unique value in the 'Budget (in Million USD)' column
print(anime_df['Budget (in Million USD)'].value_counts())
```
